**zipcode** - Индекс дома<br>
**lat** - Широта<br>
**long** - Долгота<br>
**sqft_living15** - Кв.м жилой площади у 15 соседей<br>
**sqft_lot15** - Кв.м общей площади у 15 соседей<br>

Создать новый признак price_per_sq_lot, который будет содержать среднюю стоимость за один кв. метр общей площади

```python
df['price_per_sq_lot'] = df['price']/df['sqft_lot']
```

Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома.<br>
Если реновации дома не было, то в новом признаке поставьте 0.

```python
year = []
for i, j in df.iterrows():
    if j['yr_renovated']:
        year.append(j['yr_renovated']-j['yr_built'])
    else:
        year.append(0)
```

```python
df['delta_renovated']=year
```

```python
df[df['yr_renovated'] > 0][['yr_built', 'yr_renovated', 'delta_renovated']].sample(10)
```

Создайте признаки года продажи, месяца продажи

```python
df['year_sale'] = df['date'].apply(lambda x: int(x[:4]))
df['month_sale'] = df['date'].apply(lambda x: int(x[4:6]))
```

```python
df[['date', 'year_sale', 'month_sale']].head()
```

Удалите признаки date, zipcode, lat, long

### Составьте несколько сводных таблиц
Найдите среднюю стоимость домов в зависимости от количества спален<br>
Отсортируйте от меньшей стоимости к большей

```python
df.groupby('bedrooms').agg({'price': 'mean'}).sort_values('price')
```

Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df.groupby('condition').agg({'price': ['min', 'mean', 'max']})
```

Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
pd.crosstab(index = df['waterfront'], columns = df['view'])
```

Каких домов в зависимости от этажности и количества спален больше?

```python
pd.crosstab(index = df['bedrooms'], columns = df['floors'])
```

Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома
